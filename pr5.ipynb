# Сам датасет Аниме

```python
import pandas as pd
anime = pd.read_csv('anime.csv')

anime
```

# Размерность массива - 2, т.е. всего два измерения - по X и по Y

```python
import numpy as np
print(f"anime.ndim = {anime.ndim}")
```

# Здесь выводятся наименования всех колонок. Anime_id - это уникальные номера анимешек; Name - само название аниме; Genre - жанр; Type - тип; Episodes - кол-во эпизодов; Rating - рейтинг и Members - поклонники.

```python
anime.columns
```

# Здесь через nunique() выводится кол-во уникальных значений в каждом столбце. А через []unique() можно выписать все уникальные значения в столбце

# Сортировка по определенному параметру. Мы сортируем столбец Rating в порядке убывания при помощи ascending=False. Теперь у нас датасет будет выведен от наибольшего рейтинга к меньшему

```python
anime.sort_values(by='rating', ascending=False) 
```

# Удаление всего столбца Members осуществляется при помощи команды drop(columns =[]). Теперь датасет будет выведен без стобца members

```python
anime.drop(columns = ['members'])
```

# Изменяем значения NaN в столбце Rating на Unknown и выводим только те, где rating будет равен новому значению

```python
anime=anime.replace({'rating':np.nan}, 'Unknown')
anime[anime['rating'] == 'Unknown']
```

# Ниже при помощи .drop_duplicates мы оставляем только первое появление строки, а остальные удаляются 

```python
anime.drop_duplicates()
```

# .info отображает некоторую информацию о датасете: кол-во непустых значений в каждом столбце(Non-Null Count), затем тип данных каждого столбца(Dtype), общее кол-во столбцов(total 7 columns) и кол-во использованной памяти(memory usage)

```python
anime.info()
```

# .describe возвращает основные статические показатели для числового столбца members: кол-во непустых значений(count), среднее арифметическое(mean), стандартное отклонение(std), минимальное значение(min), первый квартиль(25%), медиана(50%), третий квартиль(75%) и максимальное значение(max)

```python
anime.describe()
```

# .loc используется для быстрого доступа к данным. В 1-ом примере мы выводим всю инфу о анимешке под индексом 15. Во 2-ом для 0, 1, 2 индексов мы выводим инфу Name, Genre и Rating. В 3-ем мы выводим всю инфу где Genre = Drama

```python
row_data = anime.loc[15]
print (row_data)
```

```python
short_info = anime.loc[[0, 1, 2], ['name', 'genre', 'rating']]
print (short_info)
```

```python
drama_anime = anime.loc[anime['genre'] == 'Drama']
print (drama_anime)
```

# .to_csv сохраняет новый датасет в файл CSV

```python
anime.to_csv("anime_clear.csv", index=False)
```
